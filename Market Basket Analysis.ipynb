```python
import tkinter as t
window=t.Tk()
window.title("Billing and Recommendation")
window.geometry('800x600')

row=1

l1=t.Label(window,text="ML - Market Basket Analysis", font = 'papyrus')
l1.grid(column=2,row=row,columnspan=8)


from datetime import datetime

date=datetime.now()
ldate=t.Label(window,text=date.strftime("Date : %d - %m - %Y"))
ldate.grid(column=1,row=row)

ltime=t.Label(window,text=date.strftime(" Time : %H : %M : %S"))
ltime.grid(column=5,row=row,columnspan=13)


row+=1

l2=t.Label(window,text="Customer Details", font ="papyrus")
l2.grid(column=2,row=row)




row+=1



c_name=''
c_pn=''
c_eid=''



l11=t.Label(window,text="Name", font = 'papyrus')
l11.grid(column=1,row=row)
txt11 = t.Entry(window,textvariable="Name",width=20)
txt11.grid(column=2,row=row)

l12=t.Label(window,text="Phone Number", font = 'papyrus')
l12.grid(column=3,row=row)
txt12 = t.Entry(window,textvariable="Phno",width=20)
txt12.grid(column=4,row=row)

l13=t.Label(window,text="Email ID", font = 'papyrus')
l13.grid(column=5,row=row)
txt13 = t.Entry(window,textvariable="E ID",width=20)
txt13.grid(column=6,row=row)


def clickupdate():
    l111=t.Label(text="Name : " + txt11.get())
    c_name=''+txt11.get()
    l111.grid(column=2,row=row+100)
    l121=t.Label(text="Phone Number : " + txt12.get())
    c_pn=''+txt12.get()
    l121.grid(column=6,row=row+100)
    l131=t.Label(text="Email ID : " + txt13.get())
    c_eid=''+txt13.get()
    l131.grid(column=9,row=row+100)
    

bt13=t.Button(window,text='Update',command=clickupdate)
bt13.grid(column=10, row=row)
row+=2




l_o_p=[]
import pandas
df=pandas.read_csv(r"C:\Users\spand\Desktop\PROJECT\Goodslist.csv")
l_o_p=df.iloc[:,:1].values
prices=df.iloc[:,1:].values
lop1=[]
for i in l_o_p:
    for j in i:
        lop1.append(j)
        
prices1=[]
for i in prices:
    for j in i:
        prices1.append(j)
        
curpro=[]
TOTAL=0
##product name head, price, quantity, sub total heads
from tkinter import ttk
def clicked2():
    global row
    l2t1=t.Label(window,text="Product Name")
    l2t1.grid(column=1,row=row)
    CBT=ttk.Combobox(window,values=lop1)
    CBT.grid(column=2,row=row)
    
    
    l2t2=t.Label(window,text="Price = " )# add retrieved price from csv
    l2t2.grid(column=3,row=row)
    l2t3=t.Label(window,text="Quantity")
    l2t3.grid(column=4,row=row)
    et=t.Entry(window,width=10)
    et.grid(column=5,row=row)
    l2t4=t.Label(window,text="SubTotal = ") #price*quantity
    l2t4.grid(column=6,row=row)
    def clickedupdate():
        global curpro
        curpro.append(CBT.get())
        a=0
        temama=0
        for i in lop1:
            if i==(CBT.get()):
                temama=prices1[a]
                break
            a+=1    
        l2t2.configure(text="Price = "+ str(temama))
        temama2=et.get()
        l2t4.configure(text="Subtotal = "+str(int(temama2)*int(temama)))
        global TOTAL
        TOTAL+=int(temama2)*int(temama)
        
    bt3=t.Button(window,text='Update',command=clickedupdate)
    bt3.grid(column=8, row=row)
    row+=1
    
    
bt2=t.Button(window,text='Add product',command=clicked2)
bt2.grid(column=3,row=row-1)




def clickedsub():
    
    tem=pandas.read_csv(r"C:\Users\spand\Desktop\PROJECT\Reciepts.csv",header=None)
    tem=tem.values
    t1=len(tem)
    t2=len(tem[0])

        
    purch_no=int(tem[t1-1][0])+1
    c_name=''+txt11.get()
    c_pn=''+txt12.get()
    c_eid=''+txt13.get()
    
    tem=[purch_no,c_name,c_pn,c_eid,TOTAL,date.strftime("%d - %m - %Y"),date.strftime("%H : %M : %S")]
    import csv
    with open(r"C:\Users\spand\Desktop\PROJECT\Reciepts.csv",'a') as reciepts:
        writer=csv.writer(reciepts)
        writer.writerow(tem)
    with open(r"C:\Users\spand\Desktop\PROJECT\Products_bought.csv",'a') as PB:
        writer=csv.writer(PB)
        tem=[]
        tem.append(purch_no)
        for i in curpro:
            tem.append(i)
        writer.writerow(tem)
        
    subwindow=t.Tk()
    subwindow.title('Sub Status')
    subwindow.geometry('400x300')
    
    SL=t.Label(subwindow,text="Submitted Successfully\nTotal = "+str(TOTAL), font = 'papyrus')
    SL.grid(column=2,row=row,columnspan=8)
    
    
btsub=t.Button(window,text='Submit',command = clickedsub)
btsub.grid(column=5, row=row-1)

def clickRecom():
    import pandas
    global row
    df=pandas.read_csv(r"C:\Users\spand\Desktop\PROJECT\store data.csv",header=None)
    df=df.values

    allpr=[]
    for i in df:
        temp=[]
        for j in i:
            if type(j)==type('sda'):
                temp.append(j)
            else:
                break
        allpr.append(temp)

    df=(allpr)

    from mlxtend.preprocessing import TransactionEncoder

    te=TransactionEncoder()
    te_ary=te.fit_transform(df)

    df=pandas.DataFrame(te_ary,columns=te.columns_)
    from mlxtend.frequent_patterns import apriori
    freq_isets=apriori(df,min_support=0.003,use_colnames=True)

    freq_isets['length']=freq_isets['itemsets'].apply(lambda x: len(x))

    t=0.01
    maxrecom=50
    while(len((freq_isets[(freq_isets['length']>1) & (freq_isets['support']>t)]).values)>maxrecom):
        t+=0.001
    recom=[]

    k=((freq_isets[(freq_isets['length']>1) & (freq_isets['support']>t)])).values
    k=k.tolist()
    k.sort(key=lambda x:x[0])
    k.reverse()

    r=[]
    for i in range(len(k)):
        ta=[]
        for j in range(1,len(k[i])-1):
            ta.append(k[i][j])
        r.append(ta)

    r2=[]
    for i in r:
        for j in i:
            r2.append(list(j))

        
    while(maxrecom>1):
        for i in curpro:
            for j in r2:
                maxrecom-=1
                for k in j:
                    if i==k:
                        recom.append(j)
                    

    rrow=1
    import tkinter
    recwindow=tkinter.Tk()
    recwindow.title('Recommendations')
    recwindow.geometry('400x300')
    RL1=tkinter.Label(recwindow,text="Recommendations", font = 'papyrus')
    RL1.grid(column=2,row=rrow,columnspan=8)
    rrow+=1
    RL2=tkinter.Label(recwindow,text="Select One Offer", font = 'papyrus')
    RL2.grid(column=2,row=rrow,columnspan=8)
    rrow+=1
    from tkinter import ttk
    Rcb=ttk.Combobox(recwindow,values=recom)
    Rcb.grid(column=2,row=rrow,columnspan=8)
    rrow+=1
    def RecCon():
        temp=Rcb.get()
        offpri=0
        temp=temp.replace('}',' ')
        temp=temp.split(' ')
        for i in range(len(temp)):
            if len(temp[i])<1:
                break
            if temp[i][0]=='[':
                temp[i]=temp[i].replace('[','')
                temp[i]=temp[i].replace(']',' ')    
            
        for i in range(len(temp)):
            if len(temp[i])<1:
                break
            if temp[i][0]=='[':
                temp[i]=temp[i].replace('{','')
                temp[i]=temp[i].replace('}',' ')
           
        print(temp)
        for i in temp:
            for j in range(len(lop1)):
                if lop1[j]==i:
                    offpri+=prices1[j]
    
        
        recrecwin=tkinter.Tk()
        recrecwin.title('Offer Applied')
        recrecwin.geometry('400x150')
        rrwl1=tkinter.Label(recrecwin,text="Discount-10%  \nOriginal price - "+str(offpri)+"\nNew Price - "+str(0.9*offpri))
        rrwl1.grid(column=1,row=1)
        
        global row
        row+=1
        l2t1=tkinter.Label(window,text="Combo - "+str(temp[0]) + " + "+str(temp[1]))
        l2t1.grid(column=1,row=row)
    
        l2t2=tkinter.Label(window,text="Price = " + str(0.9*offpri))# add retrieved price from csv
        l2t2.grid(column=3,row=row)
        l2t3=tkinter.Label(window,text="Quantity = 1")
        l2t3.grid(column=4,row=row)
        l2t4=tkinter.Label(window,text="SubTotal = "+str(0.9*offpri)) #price*quantity
        l2t4.grid(column=6,row=row)
        global TOTAL
        TOTAL+=0.9*offpri
        row+=1
        
        
        
    btRecCon=tkinter.Button(recwindow,text='Confirm',command=RecCon)
    btRecCon.grid(column=3,row=rrow)
    
btRecom=t.Button(window,text='Recommend',command=clickRecom)
btRecom.grid(column=6, row=row-1)

row+=1
def clicksearch():
    import tkinter
    from tkinter import ttk
    cserwin=tkinter.Tk()
    cserwin.title('Search window')
    cserwin.geometry('500x300')
    lab=tkinter.Label(cserwin,text="Search by:")
    lab.grid(column=1,row=1)
    cb=ttk.Combobox(cserwin,values=['Name','Date','Purchase Number'])
    cb.grid(column=2,row=1)
    ent=tkinter.Entry(cserwin)
    ent.grid(column=3,row=1)
    
    def serres():
        df=pandas.read_csv(r"C:\Users\spand\Desktop\PROJECT\Reciepts.csv")
        df=df.values.tolist()
        df2=[]
        temp=0
        for i in range(len(df)):
            if i%2==0:
                df2.append(df[i])
        import tkinter
        if cb.get()=='Name':
            for i in df2:
                if i[1]==ent.get():
                    i=i[:5]
                    serlab=tkinter.Label(cserwin,text=str(i))
                    serlab.grid(column=1,row=3)
                    temp=i[0]
                    
        if cb.get()=='Date':
            for i in df2:
                if i[5]==ent.get():
                    i=i[:5]
                    serlab=tkinter.Label(cserwin,text=str(i))
                    serlab.grid(column=1,row=3)
                    temp=i[0]
                    
        if cb.get()=='Purchase Number':
            for i in df2:
                if i[0]==ent.get():
                    i=i[:5]
                    serlab=tkinter.Label(cserwin,text=str(i))
                    serlab.grid(column=1,row=3)
                    temp=i[0]
                    
        df=pandas.read_csv(r"C:\Users\spand\Desktop\PROJECT\Products_bought.csv",header=None)
        df=df.values.tolist()
        for i in df:
            if int(i[0])==int(temp):
                serlab2=tkinter.Label(cserwin,text=str(i))
                serlab2.grid(column=1,row=3)
        
    btser2=tkinter.Button(cserwin,text="commence search",command=serres)
    btser2.grid(column=2,row=2)
    
    
    
btsearch=t.Button(window,text='Search',command=clicksearch)
btsearch.grid(column=4,row=row-2)


window.mainloop()
```
